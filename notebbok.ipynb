```python
#in case the text come in multiple-line
def single_line_text(text):
    
    single_line_text = ''.join(text.split('\n'))
    return single_line_text


def input_word():
   
    word = input('Insert a word you wanted to find: ').split()
    return word

def word_count(text):
    word_count = {}
    split_text = text.split()
    for word in split_text:
        if word not in word_count:
            word_count[word] = 1
        else:
            word_count[word] += 1
    
    return word_count

def find_specific_word(dic, words_to_find):
    found_words = {}
    for word in words_to_find:
        if word in dic:
            found_words[word] = dic[word]
    
    if found_words:
        return found_words
    else:
        return f'None of the words: {words_to_find} were found in the text.'
```

```python
def variable():
    var = []
    fc = int(input('Insert your fixed cost here: '))
    sp = int(input('Insert your selling price here: '))
    vc = int(input('Insert your variable cost here: '))

    var = (fc, sp, vc)  # Directly creating a tuple
    return var

def break_even_point(fc, sp, vc):
    bep = round(fc / (sp - vc))
    return f'You need to sell the good up to {bep} units to reach the break-even point.'

# Getting inputs
inputs = variable()
#Unpacking the tuple and calling the function
print(break_even_point(*inputs))
```

```python
from langchain.text_splitter import CharacterTextSplitter
def pdfreader_text_splitter(pdf):
    pdfreader = PdfReader(pdf)

    raw_text = ''
    for i, page in enumerate(pdfreader.pages):
        content = page.extract_text()
        if content:
            raw_text += content
        else:
            print(f'There is no text in this pdf: {pdf}')
    
    # split the text using charazter Text Split such that it should not increse token
    text_splitter = CharacterTextSplitter(
        separator="\n",
        chunk_size=1000,
        chunk_overlap=200,
        length_function=len  # corrected typo here
    )

    # Split the text
    texts = text_splitter.split_text(raw_text)
    return(texts)
```
